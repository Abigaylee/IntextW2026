# Run all registered ML pipelines

Executes every pipeline in `pipeline_kit.PIPELINE_REGISTRY` with dual-track models and writes metrics/coefficients/importances to `ml-pipelines/artifacts/`.

**Run with cwd = `ml-pipelines/`.**

```python
import sys
from pathlib import Path

NOTEBOOK_DIR = Path.cwd().resolve()
if str(NOTEBOOK_DIR) not in sys.path:
    sys.path.insert(0, str(NOTEBOOK_DIR))

from pipeline_kit import PIPELINE_REGISTRY, run_pipeline_by_name

for name in sorted(PIPELINE_REGISTRY.keys()):
    print("\n" + "=" * 60)
    run_pipeline_by_name(name, seed=42, save_artifacts=True)

print("\nDone. All pipelines finished.")
```
